#### Funci√≥n

```python
def places(busqueda):
    places_result_x = gmaps.places(busqueda)
    try:
        place_id_x = places_result_x['results'][0]['place_id']
        place_x = gmaps.place(place_id = place_id_x)
    except IndexError:
        place_x = []
    return places_result_x, place_x
```
